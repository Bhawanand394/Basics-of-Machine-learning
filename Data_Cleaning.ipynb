```python
import re as text_1
text = "Hello!!! This is an Example, with Numbers 123 and Symbols @#."
```

```python
text = text.lower()
```

```python
text = text_1.sub(r'[^a-z\s]', '', text)
```

```python
text = text_1.sub(r'\s+', ' ', text).strip()
```

```python
print(text)
```

```python
tokens = text.split()
```

```python
print(tokens)
```

```python
stop_words = {"is", "am", "are", "the", "this", "and", "a", "an", "with", "to", "of", "in"}
```

```python
filtered_tokens = [word for word in tokens if word not in stop_words]
```

```python
print(filtered_tokens)
```

```python
def rule_based_stem(word):
    if word.endswith("ing"):
        return word[:-3]
    elif word.endswith("ed"):
        return word[:-2]
    elif word.endswith("s"):
        return word[:-1]
    return word
```

```python
stemmed_words = [rule_based_stem(word) for word in filtered_tokens]
```

```python
print(stemmed_words)
```
